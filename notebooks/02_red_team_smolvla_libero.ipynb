Use the LIBERO-fine-tuned checkpoint (`lerobot/smolvla_base` is **not** LIBERO-compatible). Always include the `none` baseline so each attack rate is read against its benign control.

```python
import os

os.environ['PROVAEL_INTEGRATION'] = '1'
os.environ['MUJOCO_GL'] = 'egl'
os.environ['PYOPENGL_PLATFORM'] = 'egl'

!provael attack --policy smolvla --suite libero \
    --model HuggingFaceVLA/smolvla_libero \
    --attacks none,instruction,visual,injection \
    --seeds 10 --horizon 280 --seed 0 --out runs/smolvla_libero
```

Real-model inference is seeded but **not** byte-deterministic. Read each attack rate against the `none` control; `n = 10` is a screen — report the 95% CIs. See [examples/02-redteam-smolvla-libero](https://github.com/provael/provael/tree/main/examples/02-redteam-smolvla-libero).
