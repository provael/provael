**Provael** red-teams open Vision-Language-Action (VLA) robot policies in simulation and reports an **Attack Success Rate (ASR)**.

This notebook runs the whole engine on a **CPU** — no GPU, no model download — using a deterministic stub policy and suite. Run the cells top to bottom.

## 1. Install

```python
%pip install -q provael
!provael version
```

## 2. Run your first scan

`--recipe full-sweep` runs all four attack families (instruction / visual / injection / action) for 10 episodes against the CPU stub. It finishes in well under a second and is fully deterministic.

```python
!provael attack --recipe full-sweep --out runs/first-scan
```

Each attack is tagged with its [Embodied AI Security Top 10](https://github.com/provael/provael/blob/main/docs/TOP10.md) risk (EAI01/02/04/05) and a 95% Wilson confidence interval. The `none` baseline scores 0% — every success is *lift* contributed by an attack.

## 3. Read the report

```python
import json
from pathlib import Path

!provael report --in runs/first-scan

# The machine-readable report is byte-deterministic JSON:
report = json.loads(Path('runs/first-scan/report.json').read_text())
print('overall ASR:', report['successes'], '/', report['attempts'])
```

## 4. Explore more

- `!provael list-recipes` — other named presets (`quick`, `instruction-only`, `ci-gate`).
- `!provael list-attacks` — the nine attacks across four families.
- `!provael attack --recipe full-sweep --format sarif --sarif-out provael.sarif` — emit SARIF for GitHub code scanning.
