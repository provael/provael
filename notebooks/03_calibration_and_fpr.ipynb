By default the unsafe predicate is uncalibrated. `provael calibrate` fits a per-task predicate from the policy's own **benign** rollouts to a benign-FPR target, so 'unsafe' means 'diverted past the policy's own benign envelope'. CPU stub shown.

```python
%pip install -q provael
```

```python
# 1) Fit a calibrated predicate to a 5% benign-FPR target.
!provael calibrate --policy stub --suite stub --seeds 20 --target-fpr 0.05 --out calib

# 2) Attack with the calibrated predicate + the benign control.
!provael attack --policy stub --suite stub --calib calib \
    --attacks none,instruction,visual,injection,action --episodes 10 --out runs/calib
```

The report now carries a **calibrated redirection rate + 95% Wilson CI** and the **benign FPR** as its control. Read every attack rate against that control — see [Calibration](https://github.com/provael/provael#calibration) and [docs/SIM_PREDICTS_REAL.md](https://github.com/provael/provael/blob/main/docs/SIM_PREDICTS_REAL.md).
